### TCP RTT nach Folie, Punkte generiert mit Normalverteilung ~ $N$($\mu$, $\sigma^{2}$)

```python
import numpy as np
from scipy.stats import norm

import plotly.graph_objects as go
```

```python
def calc_est_rtt(sample_rtt):
    alpha = .125
    # Initialize estimated rtt with first value of sample rtt
    estimated_rtt = [sample_rtt[0]]
    for value in sample_rtt:
        e_rtt = (1 - alpha) * estimated_rtt[-1] + alpha * value
        estimated_rtt.append(e_rtt)
    return np.array(estimated_rtt[1:])
```

```python
def calc_rtt_dev(sample_rtt, est_rtt):
    beta = .25
    dev_rtt = [5]
    for s_rtt, e_rtt in zip(sample_rtt, est_rtt):
        d_rtt = (1 - beta) * dev_rtt[-1] + beta * np.abs(s_rtt - e_rtt)
        dev_rtt.append(d_rtt)
    return np.array(dev_rtt[1:])
```

```python
mu = 210
sigma = 10

safety_margin_times = 4

sample_rtt = norm.rvs(loc=mu, scale=sigma, size=200, random_state=42)
est_rtt = calc_est_rtt(sample_rtt)
timeout_interval = est_rtt + safety_margin_times * calc_rtt_dev(sample_rtt, est_rtt)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    y=sample_rtt
))
fig.add_trace(go.Scatter(
    y=est_rtt
))
fig.add_trace(go.Scatter(
    y=timeout_interval
))
```
